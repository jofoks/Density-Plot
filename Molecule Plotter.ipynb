```python
import pandas as pds
from matplotlib import pyplot as plt
import os
BASE_DIR = os.path.expanduser('~/documents')+'/Python/Molecule_Plotter/'
os.chdir(BASE_DIR)
from main import Atoms, Molecules
```

```python
Molecules['Molecular Weight (lb/mol)']
Molecules['Gas Density @ 70°F 1 atm (lb/ft3)']
x = []
y = []

for index, row in Molecules.iterrows():
    try:
        xform = float(row['Gas Density @ 70°F 1 atm (lb/ft3)'])
        yform = float(row['Molecular Weight (lb/mol)'])
        x.append(xform)
        y.append(yform)
    except:
        pass
```

```python
fig = plt.figure(figsize=(10,7))
ax1 = plt.subplot2grid((40,40), (0,0), rowspan=40, colspan=40)
mol = ax1.scatter(x, y, label='Misc')
air = ax1.scatter(0.075, 28.96, label='Air')
propane = ax1.scatter(0.1551, 58.12, label='Propane')
plt.ylabel('Mol. Weight')
plt.xlabel('Density')
plt.title('Molecule plotter')
plt.legend()
```
